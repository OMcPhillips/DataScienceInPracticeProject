```python
import pandas as pd
```

```python
athletes = pd.read_csv("athletePopulation.csv")
```

```python
athletes.rename(columns = {'Unnamed: 0': 'NOC'}, inplace=True)
```

```python
athletes.head(5)
```

```python
athletes['Medal'].fillna('DNW', inplace = True)
```

```python
athletes = athletes.drop( ['ID', 'Indicator Name'], axis=1)
```

```python
listOf = ['Age', 'Weight', 'Height']
averages = athletes.groupby('Event')[listOf].transform('mean').fillna(0)
```

```python
averages.head(5)
```

```python
athletesMissing = athletes.drop(['Age', 'Height', 'Weight'], axis=1)
```

```python
athletesClean = pd.merge(athletesMissing, averages, left_index=True, right_index=True)
```

```python
athletesClean = athletesClean.round(2)
```

```python
athletesClean.head(5)
```

```python
athleteSubset = athletesClean.loc[(athletesClean['Year'] > 1960), :]
```

```python
athleteFinal = athleteSubset.reset_index(drop=True)
```
